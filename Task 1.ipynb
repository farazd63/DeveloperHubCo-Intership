```python
plt.figure(figsize=(8, 6))
plt.scatter(df['assigned_cohort'], df['status'], color='blue', label='Data Points')
plt.title('Scatter Chart Example')
plt.xlabel('Assigned Cohort')
plt.ylabel('Status')
plt.legend()
plt.grid(False)
plt.show()
```
